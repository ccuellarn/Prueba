## Problema 1

a) Considere la funci贸n de una flor, dada por la ecuaci贸n en coordenadas polares, 

$$
r(\theta) = \frac{1}{2}(\cos{6\theta} + 1)
$$

Grafique la funci贸n de la flor, pista usar la funci贸n 

```
plt.polar
```

````python
# tu codigo aqui

theta = np.linspace(0,2*np.pi,100)
def Getr(theta):
    return ((1/2) * (np.cos(6*theta )+1))

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, Getr(theta), color='g')
plt.show
````

b) Un diferencial de la flor se puede considerar como un $\Delta A$, dado por, 

$$
\Delta A = \frac{\Delta \theta}{2}\Big[\frac{r_i+r_{i+1}}{2}\Big]^2 
$$

Considerando ese diferencial de area, halle e imprima el area de la flor tomando 500 pasos para el 谩ngulo $\theta$ 

Pista si genera un `linspace`, ponga `retstep = True`



````python
flower_area = 0
## Tu codigo aqui
theta1,h = np.linspace(0,2*np.pi, 500, retstep=True)

def area(r,r1):
    return ( (h/2) * ((Getr(r)+Getr(r1))/2)**2 )

for i in range(0,len(theta1)-1):
    flower_area += area( theta1[i],theta1[i+1])
## Codigo para verificar el resultado no modificar
flower_area
````
