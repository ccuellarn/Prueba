```python
from scipy import constants as C
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sympy as sym
import urllib.request
```

```python
def Data(l:str):
    direccion = l
    data = pd.read_csv(direccion, delimiter='\t')
    return data
```

```python
datos1= Data(""r"C:\Users\aleja\OneDrive\Documentos\Física\Moderna imágenes\Datos1_nucleo1seno.txt")

fig, ax = plt.subplots()

ax.plot(datos1['B'],datos1['I'],color='k')

#corte con y
ax.scatter(0,0.23,color='m',label=r"$M$")
#corte con x
ax.scatter(-0.063,0,color='r',label=r"$I_c$")
#punto maximo (saturacion)
ax.scatter(0.105,0.17,color='g',label=r"$B_s$")
#punto minimo (remanente)
ax.scatter(-0.108,-0.15,color='c',label=r"$B_r$")

#etiquetas
plt.annotate("(0.23)", (0,0.25),size=11)
plt.annotate("(-0.063)", (-0.056,0),size=11)
plt.annotate("(0.17)", (0.11,0.17),size=11)
plt.annotate("(-0.15)", (-0.16,-0.15),size=11)

ax.grid()
ax.yaxis.set_major_locator(plt.MultipleLocator(0.05))
ax.xaxis.set_major_locator(plt.MultipleLocator(0.04))

plt.ylim(-0.30,0.30)
plt.xlim(-0.18,0.18)
plt.ylabel(r'$\Phi_B$')
plt.xlabel(r'$I_B$')
plt.legend(loc='upper left')
plt.show()
```
