> **2.0.1 Números factoriales:**

1. Escriba una función que calcule el factorial de $n$, con $n \in \mathbb{N}$.

$$n! = n \cdot (n - 1) \cdot (n - 2) \cdot ... \cdot 1$$

Calcule los primeros 20 números factoriales.


```python
def factorial(number):
  value = 1
  if number == 0:
    value = 1
  else:
    for i in range(1, number + 1):
      value = i*value
  return value

factorials_list = []

for i in range (0, 21):
  factorials = factorial(i)
  factorials_list.append({"El factorial de {} es".format(i): factorials})
  
print(factorials_list)
```

> **2.0.8 Choques de duración finita:**

Utilice el template de la clase partícula vista en clase para simular una colisión de duración finita. Para la interacción de las dos partículas use la siguiente fuerza de contacto (Landau-Lifshitz volumen 7):

$$\vec{f}(\vec{r}_1, \vec{r}_2) = 
\left\{
\begin{array}{cl}
  K \cdot \lvert \vec{r}_1 - \vec{r}_2 \rvert^3 \cdot \hat{n} \quad \text{ si } \quad \lvert \vec{r}_1 - \vec{r}_2 \rvert < R_1 = R_2\\
  0 \quad \text{en cualquier otro caso.}
  \end{array}
\right.
$$

donde $R_1, R_2$ son los radios de las esferas y $\hat{n}$ es el vector normal al plano de contacto. Usar $K = 1 N/m^3$, $m = 1 kg$ y $R_1 = R_2 = 2 m$. Use las siguientes condiciones iniciales:

> **Bola 1:**
$$ \vec{r}_0 = (-15, 1) $$
$$ \vec{v}_0 = (10, 0) $$
$$ \vec{a}_0 = (0, 0) $$


> **Bola 2:**
$$ \vec{r}_0 = (0, -1.5) $$
$$ \vec{v}_0 = (0, 0) $$
$$ \vec{a}_0 = (0, 0) $$


1. Crear el método de clase que calcule la fuerza entre partículas en la clase Particle. Recuerde que un cuerpo no se puede aplicar fuerza a si mismo. Cada vez que se avance en el tiempo, debe reiniciar la fuerza y la aceleración a cero, y verificar las paredes (mirar video en bloque-neón).


```python
class Particle:
    def __init__(self, pi, vi, ai, t, radius, m = 1, Id = 0):
        self.dt = t[1] - t[0]
        
        self.position = pi
        self.velocity = vi
        self.acceleration = ai

        self.Ev_Position = np.zeros((len(t), len(pi)))
        self.Ev_Velocity = np.zeros_like(self.Ev_Position)
        self.Ev_Acceleration = np.zeros_like(self.Ev_Position)

        self.radius = radius
    
    def SystemEvolution(self, i):
        self.SetPosition(i)
        self.SetVelocity(i)

        self.position += (self.dt * self.velocity)
        self.velocity += (self.dt * self.acceleration)
        
    def SetPosition(self, i):
        self.Ev_Position[i] = self.position
    
    def SetVelocity(self, i):
        self.Ev_Velocity[i] = self.velocity
    
    def GetPosition(self, scale = 1):
        return self.Ev_Position[::scale]

    def GetVelocity(self, scale = 1):
        return self.Ev_Velocity[::scale]
    
    def CheckWalls(self, walls):
        for i in range(2):
            if self.position[i] + self.radius > walls[i][1] and self.velocity[i] > 0.:
                self.velocity[i] = -1.0*self.velocity[i]
            if self.position[i] - self.radius < walls[i][0] and self.velocity[i] < 0.:
                self.velocity[i] = -1.0*self.velocity[i]
    
    def CheckCollisions(self, self_):
        K = 1
        x1, y1 = self.position[0], self.position[1]
        x2, y2 = self_.position[0], self_.position[1]

        distance = np.linalg.norm(self.position - self_.position)
        n = np.array([(x2 - x1)/distance, (y2 - y1)/distance])

        if distance <= (self.radius + self_.radius):
            impact = K*(distance**3)*n
            self.acceleration = -impact
        else:
            self.acceleration = 0
    
```

```python
def RunSimulation1(t, walls):
    r0 = np.array([-15., 1.])
    v0 = np.array([10., 0.])
    a0 = np.array([0., 0])

    r1 = np.array([0., -1.5])
    v1 = np.array([0., 0.])
    a1 = np.array([0., 0])
    
    Sphere1 = Particle(r0, v0, a0,t, 2)
    Sphere2 = Particle(r1, v1 ,a1,t, 2)
    
    walls_ = walls.copy()
    
    for it in tqdm(range(len(t)), desc='Running simulation', unit=' Steps'):
        sleep(0.0001)

        Sphere1.CheckCollisions(Sphere2)
        Sphere1.SystemEvolution(it)
        Sphere1.CheckWalls(walls_)
        
        Sphere2.CheckCollisions(Sphere1)
        Sphere2.SystemEvolution(it)
        Sphere2.CheckWalls(walls_)
    
    return Sphere1, Sphere2

dt = 0.1
tmax = 10
t = np.arange(0.,tmax,dt)
scale = 1
walls = np.array([[-20., 20.],[-20., 20.]])

Sphere1, Sphere2 = RunSimulation1(t, walls)

t = t[::scale]
```

```python
fig = plt.figure(figsize = (7, 7))
ax = fig.add_subplot(111)

def init():
    ax.set_xlim(walls[0][0], walls[0][1])
    ax.set_ylim(walls[1][0], walls[1][1])
    
def Update(i):
    ax.clear()
    init()
    ax.set_title(r'$ t=%.2f \ s$' %(t[i]))
    
    # Información de la primera bola:
    x1 = Sphere1.GetPosition(scale)[i, 0]
    y1 = Sphere1.GetPosition(scale)[i, 1]
    vx1 = Sphere1.GetVelocity(scale)[i, 0]
    vy1 = Sphere1.GetVelocity(scale)[i, 1]
    circle1 = plt.Circle((x1, y1), Sphere1.radius, fill = True, color = 'k')
    ax.arrow(x1, y1, vx1*0.7, vy1*0.7, color = 'r', width = .3, head_width = 1, length_includes_head = True)
    ax.add_patch(circle1)
    
    # Información de la segunda bola:
    x2 = Sphere2.GetPosition(scale)[i, 0]
    y2 = Sphere2.GetPosition(scale)[i, 1]
    vx2 = Sphere2.GetVelocity(scale)[i, 0]
    vy2 = Sphere2.GetVelocity(scale)[i, 1]
    circle2 = plt.Circle((x2, y2), Sphere2.radius, fill = True, color = 'k')
    ax.arrow(x2, y2, vx2*0.7, vy2*0.7, color = 'r', width = .3, head_width = 1, length_includes_head = True)
    ax.add_patch(circle2)
    
Animation = anim.FuncAnimation(fig, Update, frames = len(t), init_func = init)
Animation.save("Collisions.mp4")
```

> **Campo de Velocidades:**

Calcular el campo de velocidades cerca de la superficie de un cilindro de radio $R = 2$ cm. Para esta tarea realizar los siguientes pasos:




**a)** Definir una discretización en los ejes $x$ e $y$, donde la región es: $A \in [-4, 4]$ con $25$ puntos en cada eje.

```python
x = np.linspace(-4,4,25)
y = np.linspace(-4,4,25)
```

**b)** Definir la función potencial del flujo dada por:

$$\phi(x,y) = Vx (1 - \frac{R^2}{x^2 + y^2})$$

donde V = 2 cm/s.

```python
def FlowPotential(x,y):
  return (2)*x*(1 - (4 / (x**2 + y**2)))
```

**c)** Calcule y guarde adecuadamente el campo de velocidades usando la definición de
derivada parcial central como:

$$ v_x = \frac{\partial \phi}{\partial x} $$
$$ v_y = -\frac{\partial \phi}{\partial y} $$

use h = 0.001. Note que al interior del cilindro el campo de velocidades debe ser igual
a cero.

```python
h = 0.001
def VxFlow(x,y):
  if h != 0:
    return (FlowPotential(x+h,y) - FlowPotential(x-h,y)) / (2*h)

def VyFlow(x,y):
  if h != 0:
    return -(FlowPotential(x,y+h) - FlowPotential(x,y-h)) / (2*h)
```

```python
FlowCoordinates = []

for i in x:
    for j in y:
        if (i**2 + j**2) > 4:
            xCoord = i
            yCoord = j
            vxCoord = VxFlow(i, j)
            vyCoord = VyFlow(i, j)
            FlowCoordinates.append([xCoord, yCoord, vxCoord, vyCoord])
```

>**Raíces de polinomios.**
Calcular todas las raíces reales de 
$$f(x) = 3x^5 + 5x^4 - x^3$$

```python
def Function(x):
    return (3*(x**5) + 5*(x**4) - x**3)
  
def Derivate(f,x,h=1e-4):
    return (f(x+h)-f(x-h))/(2*h)
  
def GetNewtonRaphson(f,df,xn,itmax = 1000, precision=1e-5):
    error = 1
    it = 0

    while error > precision and it <= itmax:
        try:
            xn1 = xn - f(xn)/df(f,xn)
            error = np.abs(f(xn)/df(f,xn))
        except ZeroDivisionError:
            print('Division por cero')
        it += 1
        xn = xn1
    
    if it == itmax:
        False
    else:
        return xn
      
def GetAllRoots(x,tolerancia=4):
    Roots = np.array([])
    for i in x:
        root = GetNewtonRaphson(Function,Derivate,i)
        if root != False:
            croot = np.round(root, tolerancia)
            if croot not in Roots:
                Roots = np.append(Roots,croot)
    Roots.sort()
    print ("Las raíces reales del polinomio son {}, {} y {}.".format(Roots[0], Roots[1], Roots[2]))
```
